# Logo generation notebook

The autoplotlib logo was generated with ChatGPT (4 and 3.5) using a variation of the prompt:

```
Please create a Python script that uses matplotlib to create a square logo for a tool called "autoplotlib". Make the logo look retro and dark. The background should be black or dark blue. The text should be red.
```

Some of the code was adapted from the original LM output.

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Define the color palette
dark_blue = "#17202A"
retro_green = "#27AE60"
retro_yellow = "#F1C40F"
retro_red = "#C0392B"

# Set up the plot
fig, ax = plt.subplots(figsize=(4, 4))
plt.xlim(0, 4)
plt.ylim(0, 4)
plt.axis("off")

# Add the background
rect = Rectangle((0, 0), 4, 4, linewidth=0, edgecolor=dark_blue, facecolor=dark_blue)
ax.add_patch(rect)

# Add the "autoplotlib" text
plt.text(1, 2.7, "AUTO", fontfamily="monospace", fontsize=30, color=retro_green)
plt.text(1, 1.7, "PLOTLIB", fontfamily="monospace", fontsize=30, color=retro_yellow)

# Add two lines
plt.plot([1, 3], [3.2, 3.2], color=retro_red, linewidth=3)
plt.plot([1, 3], [1.2, 1.2], color=retro_red, linewidth=3)

# Save the logo to a file
plt.savefig("autoplotlib_logo.png", dpi=300, bbox_inches="tight", pad_inches=0)

# Show the logo
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter

np.random.seed(74)

def create_logo(size, text, filename):
    fig, ax = plt.subplots(figsize=(size, size))
    ax.set_xlim(0, size)
    ax.set_ylim(0, size)
    ax.axis("off")

    # Generate background noise
    bg_noise = np.random.rand(size, size)
    bg_noise = gaussian_filter(bg_noise, sigma=2)

    # Create a retro color map
    cmap = plt.get_cmap("magma")

    # Display background noise
    ax.imshow(bg_noise, cmap=cmap, origin="upper", extent=[0, size, 0, size])

    # Add text
    ax.text(size/2, size/2, text,
            fontname="monospace",
            fontsize=75,
            color="white",
            # make bold
            fontweight="bold",
            alpha=0.8,
            ha="center",
            va="center")

    # Add grid lines
    for i in range(0, size, size//5):
        ax.axhline(i, color="white", alpha=0.1)
        ax.axvline(i, color="white", alpha=0.1)

    # Save the logo to a file
    plt.savefig(filename, dpi=300, bbox_inches="tight")

create_logo(10, "autoplotlib", "autoplotlib_logo.png")
````

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Create the canvas
fig, ax = plt.subplots(figsize=(5, 5))
ax.set_xlim(0, 100)
ax.set_ylim(0, 100)
ax.set_aspect('equal')

# Set black background

background_color = '#000040'
fig.patch.set_facecolor(background_color)
ax.set_facecolor(background_color)

# Remove axis
ax.axis('off')

# Add dark blue rectangle
rect = Rectangle((0, 0), 100, 100, facecolor=background_color, alpha=1.0)
ax.add_patch(rect)

# Draw retro lines
for x in range(0, 101, 20):
    plt.plot([x, 100-x], [0, 100], color='orange', lw=6, alpha=0.5)
    plt.plot([0, x], [100, 100-x], color='orange', lw=6, alpha=0.5)

# Add text
plt.text(50, 50, 
         'autoplotlib', 
         fontsize=38, 
         # make bold
         fontweight='bold',
         fontname='monospace',
         # add background color same as the rectangle
         bbox=dict(facecolor=background_color, alpha=1.0, 
                   # remove line of bbox
                   edgecolor=background_color,
                   boxstyle='round,pad=0.4'),
         color='lightblue', alpha=0.9, 
         ha='center', va='center')

# Save and display the logo
plt.savefig('autoplotlib_logo.png', dpi=300, bbox_inches='tight', facecolor=fig.get_facecolor())
plt.show()
````
